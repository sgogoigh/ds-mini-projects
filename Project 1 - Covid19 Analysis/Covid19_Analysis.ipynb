# Relating Variables using Scatterplots

```python
#mapping points
sns.relplot(x="cumulative_cases",y="death",data=data)
```

```python
sns.relplot(x="total_people_tested",y="death", hue ="recovered", data=data)
```

```python
#checking relations for every pair of columns
sns.pairplot(data)
```

```python
#pairing in line-plots
sns.relplot(x='total_people_tested', y='cumulative_cases',kind="line",data=data)
```

```python
sns.catplot(x='cumulative_cases',y="recovered",data=data)
```
